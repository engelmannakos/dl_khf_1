```python
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
rgbs = []
imgs = []
imgs.append(mpimg.imread('./shrek.jpg'))
imgs.append(mpimg.imread('./thomas.jpg'))
imgs.append(mpimg.imread('./anakin.jpg'))
imgs.append(mpimg.imread('./bully.jpg'))
imgs.append(mpimg.imread('./shaggy.jpg'))

for img in imgs:

    lum_img = img.reshape(-1, 3)
    red = np.array([])
    green = np.array([])
    blue = np.array([])
    for c in lum_img:
        red = np.append(red, c[0])
        green = np.append(green, c[1])
        blue = np.append(blue, c[2])

    rgbs.append(np.array([red, green, blue]))
```

```python
rgb = rgbs[0]
rmean = np.mean(rgb[0])
gmean = np.mean(rgb[1])
bmean = np.mean(rgb[2])

rstd = np.std(rgb[0])
gstd = np.std(rgb[1])
bstd = np.std(rgb[2])

red2 = (rgb[0]-rmean)/rstd
green2 = (rgb[1]-gmean)/gstd
blue2 = (rgb[2]-bmean)/bstd
print("First picture")
print("Means")
print(np.mean(red2))
print(np.mean(green2))
print(np.mean(blue2))
print("Standard deviations")
print(np.std(red2))
print(np.std(green2))
print(np.std(blue2))
print("--------------------")
```

```python
import wave

wav = wave.open("famguy.wav", "r")
peter = wav.readframes(-1)
peter = np.frombuffer(peter, "int16")

wav2 = wave.open("Michael_Scott.wav", "r")
michael = wav2.readframes(-1)
michael = np.frombuffer(michael, "int16")

petermean = np.mean(peter)
peterstd = np.std(peter)
peter2 = (peter - petermean)/peterstd

michaelmean = np.mean(michael)
michaelstd = np.std(michael)
michael2 = (michael - michaelmean)/michaelstd

print("Family Guy wav")
print(np.mean(peter2))
print(np.std(peter2))

print("--------------------")
print("The Office wav")

print(np.mean(michael2))
print(np.std(michael2))
```
